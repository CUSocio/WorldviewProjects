#### The Puzzle
My focus for research project one is on youth identity formation and alcohol use as gleaned through participation in four areas of social life relevant for identity formation: family, extra-curricular activities, education, and alcohol use. Identity scholars across many disciplines identify the first three areas as incredibly relevant for forming culturally valued identities that contribute to positive health outcomes across the life course. However, absence of such valued identities is linked to formation of stigmatized identities, such as that of a drinker. My previous qualitative work on drinker identity underscores that individuals who lack a sense of belongingness in two or more of these areas often seek perceived belongingness or comradery among drinkers. While some individiuals go on to identify as students, family members, or athletes, those who do not are at risk of identifying as a drinker. When identification as a drinker occurs, bonds are forged through drinking cultures, fostering additional consumption which leads to alcohol misuse. This process theoretically begins in childhood. Therefore, I seek answers to the general question: 'does a lack of participation in the family, extra-curricular activites, or education contribute to adolescent alcohol use?"

#### Why? As qualitative work helps to articulate processes, grounding qualitative findings in quantitative data offers additional support for conclusions. For these reasons, among others, I seek answers to the above research question.

Hypothesis 1: When one dimension of self is missing, we will see an increase in alcohol use.

Hypothesis 2: Lower perceptions of familial-based identity will result in increased consumption, on average.

#### Materials Used
The data were made available by UCI Machine Learning and were made accessible through @data.world.com. The data is also available at kaggle.com/datasets/uciml/student-alcoho-consumption. The data were collected by P. Cortez and A. Silva from a math course in Portugal in 2008. I collected the data by downloading the file "student-mat.csv" from data.world.com. 382 students were sampled and included in the dataset. There are 33 columns in the dataset that relate to information on age, parental education, gender, family size, parental occupation, and binary questions related to extra educational support (beyond family), family support, extra-curricular activities, and alcohol consumption (Likert Scale from 1-5; on both workdays and on the weekends; will make a composite measure). There are additional questions about internet usage, romantic relationships, school absences, and longitudinal questions about students' grades at three time periods. 

#### What does the data represent?
I am leveraging this data set as a theoretical representation of four portions of self: educational identity, extra-curricular identity, family-based identity, and drinker identity. For educational identity, a desire to pursue higher education is representative of an educational identity. Though many individuals go on to attempt higher educational degrees and then form drinker identities instead, given the data available, this is the most accessible source. Ideally, the students would be followed throughout their degrees, but this data is not available. A variable on family relations is representative of a family-based identity. If individuals find value in their family dynamics at home, they likely put value in that component of self, which is indicative of a valued identity. I will use the variable related to extra-curricular activities as indicative of another valued identity. Those that take part in exta-curricular activities form identities related to those activities. The last variable is related to alcohol use, and theoretically tied to a budding identity as a drinker. 

#### Variables Used in Analysis
In this analysis, I use variables "famrel" (i.e., percieved quality of family relationship; Likert scale from 1-5), "activities" (i.e., exta-curricular activities; binary: yes/no response), "higher" (i.e., desire to take higher education; binary: yes/no response), and "wkalc" (i.e., a composite measure I will create which combines the two measures of alcohol consumption). There are additionally two variables which may provide additional information regarding alcohol use and the formation of an educational identity, and, depending upon the findings, I may investigate both "famsup" (i.e., family educational support) and "schoolsup" (i.e., extra educational support). Both are binary, yes/no, measures. 

The final two variables may highlight how individuals may validate particular aspects of self of adolescents, guiding them away from substance use.

```python
# creating the composite variable "wkalc"
df['wkalc'] = df['Dalc'] + df['Walc']
df.head()
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

sns.boxplot(data=df, x="famrel", y="wkalc", palette = "mako", notch=True)

plt.title('Family-based Identity and Alcohol Use', fontsize=20, family='serif', weight ='bold')
plt.xlabel("Strength of Family-based Identity", fontsize=14, family = 'serif')
plt.xticks(fontsize=12, family='serif')
plt.ylabel('Drinks/week', fontsize=14, family='serif')
plt.yticks(fontsize=12, family='serif')

ax.yaxis.grid(True)
sns.despine(trim=True, left=True)

plt.show()
```

```python
fig, ax = plt.subplots()

sns.boxplot(data=df, x='higher', y="wkalc", palette = 'viridis', notch=True)

plt.title('Educational Identity and Alcohol Use', fontsize=20, family='serif', weight ='bold')
plt.xlabel("Presence of Educational Identity: 0=No, 1=Yes", fontsize=14, family = 'serif')
plt.xticks(fontsize=12, family='serif')
plt.ylabel('Drinks/week', fontsize=14, family='serif')
plt.yticks(fontsize=12, family='serif')

ax.yaxis.grid(True)
sns.despine(trim=True, left=True)

plt.show()

fig, ax = plt.subplots()

sns.boxplot(data=df, x='activities', y="wkalc", palette = 'viridis', notch=True)

plt.title('Extra-curricular Identity and Alcohol Use', fontsize=20, family='serif', weight ='bold')
plt.xlabel("Presence of Extra-curricular Identity: 0=No, 1=Yes", fontsize=14, family = 'serif')
plt.xticks(fontsize=12, family='serif')
plt.ylabel('Drinks/week', fontsize=14, family='serif')
plt.yticks(fontsize=12, family='serif')

ax.yaxis.grid(True)
sns.despine(trim=True, left=True)

plt.show()
```

#### Interpretation
Given the outputs above, I find evidence that lacking either familial or educational based dimensions of self contribute to increased alcohol use. However, for extra-curricular identity, there is absolutely no correlation. There is evidence in support of both Hypothesis 1 and 2. 

As familial-based identity was a categorical measure based on percieved strength of relation, points of interest were revealed. One such point is that having some form of ambiguity around the relationship, such as a score of three would indicate, also contributes to increased alcohol use on average. With average weekly drinks sitting at 4 for a score of three, I assume that the relationship is a significant increase. What is most interesting, however, is that lacking any sense of familial identity, such as a score of one would indicate, reduced consumption (3.5 drinks/week on average) when compared to scores of two (4.5 drinks/week on average) and three. This could be the result of many factors, such as acceptance of familial relations, therefore reducing a drive to cope through alcohol use, or maybe from a lack of family relations at all. 

Educational identity was strongly associated with with increased consumption; lacking an identity as a student fostered additional consumption (i.e., 2 more drinks/week on average). This is an interesting finding, and supports much of the past research on drinker identity formation. 

There are many avenues which future research could take. Given my inability to look for strong associations, future research could look at a logistic regression between the presence of valued identities and drinker identity. Furthermore, a follow up investigation as to whether or not individuals gained a sense of belonging from their identification with culturally valued identities could produce an answer as to why we see some sort of association between educational and family-based components of self and alcohol use. Also, investigating why extra-curricular identity did not relate to alcohol consumption at all could prove interesting. Maybe some extra-curriculars contributed while others did not. Lastly, though it was not included in this investigation, considering how alcohol use varied by gender and depended upon these other aspects of self might reveal some interesting findings. There is recent evidence that there is a gender shift (i.e., women are consuming more than men on average) occuring in consumption of alcohol and alcohol culture. Finding data related to contemporary drinking culture may reveal why this transition is taking place.
